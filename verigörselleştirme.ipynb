```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
data = pd.read_csv('Covid19-Turkey.csv', index_col = 'Date')
```

```python
data.head()
```

```python
data.tail()
```

```python

plt.figure(figsize=(40,8))
sns.lineplot(data = data['Daily Cases'], label = 'Daily Cases')
sns.lineplot(data = data['Active Cases'], label = 'Active Cases')
plt.xticks(rotation=90)
plt.title('Daily and Active Cases')
plt.xlabel('Timeline')
plt.ylabel('Cases')
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(x = data['Total Cases']-data['Active Cases'], y = data['Total Deaths'], hue = data['Total Cases'])
plt.title('Deaths - Healing cases with Total Cases')
plt.xlabel('Healing')
plt.ylabel('Deaths')
plt.show()
```

```python
data_july = data.loc["7/1/2020":"7/31/2020"]
plt.figure(figsize=(32,6))
sns.barplot(x = data_july.index, y = data_july['Daily Cases'])
plt.title('Daily Cases on July')
plt.xlabel('Timeline')
plt.ylabel('Cases')
plt.show()
```

```python
plt.figure(figsize=(10,6))
data_june = data.loc["6/1/2020":"6/30/2020"]
sns.scatterplot(x = data_june['Total Cases']-data_june['Active Cases'], y = data_june['Total Deaths'], hue = data_june['Total Cases'])
plt.title('Deaths - Healing cases with Total Cases on June')
plt.xlabel('Healing')
plt.ylabel('Deaths')
plt.show()
```

```python
sns.jointplot(x=data['Total Cases'], y=data['Active Cases'], hue_order = data.index, kind="kde", shade=True, height = 10)
plt.title('Total-Active Cases')
plt.xlabel('Total Cases')
plt.ylabel('Active Cases')
plt.show()
```
